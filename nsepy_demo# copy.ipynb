```python
# Get daily stock price by nse history
from nsepy.history import get_price_list
from nsepy.history import get_history
from datetime import date
import pandas as pd
prices = get_price_list(dt=date(2021,1,25))

for index, rows in prices.iterrows():
    if index == 0:
        symbol_price = get_history(symbol=rows['SYMBOL'],
                    start=date(2021,1,1),
                    end=date(2021,1,26))

        symbol_price.to_csv('prices_daily_260121.csv', sep=',' , encoding='utf-8', index=True)
    else: # index > 0 and index <= 10:
         symbol_price = get_history(symbol=rows['SYMBOL'],
                    start=date(2021,1,1),
                    end=date(2021,1,26))
         symbol_price.to_csv('prices_daily.csv', mode='a', sep=',' , encoding='utf-8', index=True, header=False)
  #  else:
   #     break
```
